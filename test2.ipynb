```python
import re
def bq_parser(text):
    res = ''
    open_tag = '<blockquote style="border: 1px dashed grey; border-left: 3px solid #4188f1; padding:2px; margin:5px;">'
    close_tag = '</blockquote>'
    text_wo_bq = re.sub(r'>(.*?)\n', r'\1\n', text) # 맨 앞 >기호 없애기
    idx = 0 #첫 인덱스
    tmp = '' # blockquote로 묶일 수 있는 부분인지 확인할 것.
    while idx <len(text_wo_bq):
        # idx 맨 처음부터 빈 경우
        if re.match('>(.*?)\n', text_wo_bq[idx:]):
            tmp += re.match('>(.*?)\n', text_wo_bq[idx:]).group(0)
            idx += len(re.match('>(.*?)\n', text_wo_bq[idx:]).group(0))
        # 나머지 경우 tmp 부분은 bq_parser로 처리한 후에
        else:
            res += bq_parser(tmp)
            idx1 = len(re.match('.*\n(>|^)', text_wo_bq[idx:]).group(0))
            res += text_wo_bq[idx:idx+idx1]
            idx += idx1

    return res
```

```python
def bq_parser(text):
    res = ''
    open_tag = '<blockquote style="border: 1px dashed grey; border-left: 3px solid #4188f1; padding:2px; margin:5px;">'
    close_tag = '</blockquote>'
    text_wo_bq = re.sub(r'>(.*?)(\n|$)', r'\1\2', text) # 맨 앞 >기호 없애기
    print(text_wo_bq)
    idx = 0 #첫 인덱스
    tmp_block = '' # blockquote로 묶을 수 있는 부분인지 확인할 것.
    tmp_etc = "" # blockquote 바깥의 부분
    while idx <len(text_wo_bq):
        # 첫 번째 개행 위치 찾기
        idx1 = text_wo_bq[idx:].find('\n')
        # 개행할 게 남아있으면
        if idx1 > -1:
            tmp_line = text_wo_bq[idx:idx + idx1 + 1]
            # 만약 여전히 블록 안에 있을 때
            if re.match(r">(.*?)\n", tmp_line):
                if tmp_etc != "":
                    print('tmp_etc:', tmp_etc)
                    res += tmp_etc
                    tmp_etc = ""
                tmp_block += tmp_line
            # blockquote 안에 없는 부분이면
            else:
                if tmp_block != "":
                    print('tmp_block:', tmp_block)
                    res += bq_parser(tmp_block[:-1])+"\n"
                    tmp_block = ""
                tmp_etc += tmp_line

            idx = idx + idx1 + 1 # 개행문자 다음에 추가
        # 마지막줄
        else:
            tmp_line = text_wo_bq[idx:]
            if re.match(r">(.*?)\n", tmp_line):
                if tmp_etc !="":
                    print('tmp_etc:', tmp_etc)
                    res += tmp_etc,
                    tmp_etc = ""
                tmp_block += tmp_line
                res += bq_parser(tmp_block[:-1])
                print('tmp_block:', tmp_block)
            else:
                if tmp_block != "":
                    print('tmp_block:', tmp_block)
                    res += bq_parser(tmp_block[:-1])+"\n"
                    tmp_block = ""
                tmp_etc += tmp_line
                res += tmp_etc
                print('tmp_etc:', tmp_etc)
            idx = len(text_wo_bq)

    return open_tag+'\n'+res+'\n'+close_tag
```
